```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('D:/Stacks/DL Frames/eigenvals.csv')
df.head()
```

```python
df['diff'] = df["Eig_1"] - df["Eig_2"]
df['trace_2'] = df["Eig_1"] + df["Eig_2"]
df['trace_3'] = df["Eig_1"] + df["Eig_2"] + df["Eig_3"] 
df['trace_4'] = df["Eig_1"] + df["Eig_2"] + df["Eig_3"] + df["Eig_4"] 
df['trace_5'] = df["Eig_1"] + df["Eig_2"] + df["Eig_3"] + df["Eig_4"] + df["Eig_5"]
df.head()
```

```python
df_131 = df.loc[df["Folder"] == 13100]
df_118 = df.loc[df["Folder"] == 11851]
df_029 = df.loc[df["Folder"] == 2987]
```

### First Eigen-value

```python
mu = df_131["Eig_1"].mean()
sig = df_131["Eig_1"].std()

x = [i for i in range(len(df_131))]

f = plt.figure()
f.set_figwidth(16)
# f.set_figheight(22)
plt.stem(x, df_131["Eig_1"])
plt.plot(x, [mu for i in range(len(x))], 'r')
plt.plot(x, [mu + sig for i in range(len(x))], 'g')
plt.plot(x, [mu - sig for i in range(len(x))], 'g')
plt.xticks(np.arange(min(x), max(x) + 1, 2.0))
plt.show()


print(f"Mean: {mu}, Sigma: {sig}")

```
